```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np   # 배열 사용할 때 numpy
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/IP/Images/girl.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gimg1, cmap='gray')
```

```python
gMul = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype=np.ubyte)
```

```python
def saturation(val):
  if val > 255:
    return 255
  elif val < 0:
    return 0
  else:
    return val
```

```python
val = 2

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    mulVal = np.int32(gimg1[h,w]) * val
    gMul[h,w] = saturation(mulVal)

#plt.imshow(gMul, cmap='gray')
#plt.imshow(gMul, cmap='gray', vmin=0, vmax=255)
cv2_imshow(gMul)
```
